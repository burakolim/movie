```python
# Kullanıcının izlediği filmleri al
user_movies = ratings[ratings['userId'] == user_index]

# Kullanıcının izlediği filmleri ve ilgili film bilgilerini birleştir
user_movies_details = user_movies.merge(movies, on='movieId', how='left')

# Kullanıcının izlediği filmleri görüntüle
print("Kullanıcının İzlediği Filmler:")
print(user_movies_details[['movieId', 'title', 'rating']])
```

```python
# Kullanıcının izlediği filmleri al
user_movies = ratings[ratings['userId'] == user_index]
num_user_movies = len(user_movies)
print(f"Kullanıcının izlediği film sayısı: {num_user_movies}")
```

```python
if num_user_movies > 0:
    user_movie_ids = user_movies['movieId'].tolist()
    user_genres = movies[movies['movieId'].isin(user_movie_ids)]['genres'].str.split('|').explode()
    user_genre_counts = user_genres.value_counts()

    print("Kullanıcının İzlediği Türler:")
    print(user_genre_counts)
     # Film türlerini işlemeye uygun hale getir
    movies['genres'] = movies['genres'].str.split('|')

    # Kullanıcının izlediği türlerden öneri yap
    recommended_movies = movies[movies['genres'].apply(lambda x: any(genre in x for genre in user_genre_counts.index))]
     # Öneri sayısını kontrol et
    if recommended_movies.empty:
        print("Eşleşen film bulunamadı.")
        top_rated_movies = ratings.groupby('movieId')['rating'].mean().sort_values(ascending=False).head(5)
        recommended_movies = movies[movies['movieId'].isin(top_rated_movies.index)]
    else:
        # 5 öneri filmi al
        recommended_movies = recommended_movies.sample(n=min(5, len(recommended_movies)))

    print("\nKullanıcıya Önerilen Filmler:")
    print(recommended_movies[['movieId', 'title', 'genres']])
else:
    print("Bu kullanıcı hiç film izlememiş.")   
```
